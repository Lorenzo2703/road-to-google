## 📚 Theory

### Arrays
An array is a contiguous block of memory storing elements of the same type. In Python, `list` acts as a dynamic array.

| Operation | Time Complexity | Notes |
|-----------|----------------|-------|
| Access by index | O(1) | Direct memory addressing |
| Search (unsorted) | O(n) | Must check every element |
| Insert at end | O(1) amortized | Resizing takes O(n) occasionally |
| Insert/Delete at index | O(n) | Elements must be shifted |

### Key Patterns for Arrays
1. **Prefix Sum**: Precompute cumulative sums to answer range queries in O(1).
2. **Kadane's Algorithm**: Find maximum subarray sum in O(n).
3. **Two Pointers**: Often used when the array is sorted (more on this on Day 03).
4. **In-place Modification**: Modify array without extra space (Google highly values space efficiency).

### Strings in Python
- Strings are **immutable** — every modification (like concatenation) creates a new string.
- Use `list(s)` to convert for in-place modifications, then `''.join(lst)`.
- String comparison is `O(min(n, m))` where `n` and `m` are the lengths.
- Common operations: `split()`, `strip()`, `lower()`, `replace()`, `find()`.

### 💡 Tips for Google Interviews
- Always clarify constraints: Is the array sorted? Are there duplicates? What about empty inputs or negative numbers?
- Always state the brute force solution quickly, then optimize.
- Consider HashMaps to trade space for time (e.g., turning an O(n²) search into O(n)).

```python
# Solution: Two Sum
def two_sum_solution(nums: list[int], target: int) -> list[int]:
    """O(n) time, O(n) space using HashMap."""
    seen = {}  # value -> index
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []
```

```python
# Solution: Product of Array Except Self
def product_except_self_solution(nums: list[int]) -> list[int]:
    """O(n) time, O(1) extra space."""
    n = len(nums)
    result = [1] * n
    
    prefix = 1
    for i in range(n):
        result[i] = prefix
        prefix *= nums[i]
    
    suffix = 1
    for i in range(n - 1, -1, -1):
        result[i] *= suffix
        suffix *= nums[i]
    
    return result
```

```python
# Solution: Longest Distinct Subarray
def longest_distinct_subarray_solution(nums: list[int]) -> int:
    """O(n) time, O(n) space using Sliding Window and Set."""
    seen = set()
    left = 0
    max_len = 0
    
    for right in range(len(nums)):
        while nums[right] in seen:
            seen.remove(nums[left])
            left += 1
        seen.add(nums[right])
        max_len = max(max_len, right - left + 1)
        
    return max_len
```
