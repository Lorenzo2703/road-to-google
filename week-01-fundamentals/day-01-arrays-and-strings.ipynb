## 📚 Theory

### Arrays

An array is a contiguous block of memory storing elements of the same type.

| Operation | Time Complexity |
|-----------|----------------|
| Access by index | O(1) |
| Search (unsorted) | O(n) |
| Insert at end | O(1) amortized |
| Insert at index | O(n) |
| Delete at index | O(n) |

### Key Patterns for Arrays

1. **Prefix Sum**: Precompute cumulative sums for range queries in O(1)
2. **Kadane's Algorithm**: Find maximum subarray sum in O(n)
3. **Dutch National Flag**: Partition array into three sections (used in quicksort)
4. **In-place Modification**: Modify array without extra space

### Strings in Python

- Strings are immutable — every modification creates a new string
- Use `list()` to convert for in-place modifications, then `''.join()`
- String comparison is O(min(n,m)) where n, m are lengths
- Common operations: `split()`, `strip()`, `lower()`, `replace()`, `find()`

### Tips for Google Interviews

- Always clarify: Is the array sorted? Are there duplicates? What about empty input?
- Think about in-place solutions first — Google values space efficiency
- Consider hash maps to trade space for time (O(n) vs O(n²))

```python
# Solution: Two Sum

def two_sum_solution(nums: list[int], target: int) -> list[int]:
    """O(n) time, O(n) space using HashMap."""
    seen = {}  # value -> index
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []  # No solution found

# Why this works:
# - For each number, we check if its complement (target - num) was seen before
# - HashMap gives O(1) lookup
# - Single pass through the array = O(n) total

print("Two Sum solution loaded ✅")
```

```python
# Solution: Valid Anagram

from collections import Counter

def is_anagram_solution(s: str, t: str) -> bool:
    """O(n) time, O(1) space (26 chars max)."""
    if len(s) != len(t):
        return False
    return Counter(s) == Counter(t)

# Alternative without Counter:
def is_anagram_manual(s: str, t: str) -> bool:
    if len(s) != len(t):
        return False
    count = {}
    for c in s:
        count[c] = count.get(c, 0) + 1
    for c in t:
        count[c] = count.get(c, 0) - 1
        if count[c] < 0:
            return False
    return True

print("Valid Anagram solution loaded ✅")
```

```python
# Solution: Product of Array Except Self

def product_except_self_solution(nums: list[int]) -> list[int]:
    """O(n) time, O(1) extra space."""
    n = len(nums)
    result = [1] * n
    
    # Left pass: result[i] = product of all elements to the left of i
    prefix = 1
    for i in range(n):
        result[i] = prefix
        prefix *= nums[i]
    
    # Right pass: multiply by product of all elements to the right of i
    suffix = 1
    for i in range(n - 1, -1, -1):
        result[i] *= suffix
        suffix *= nums[i]
    
    return result

# Trace through [1, 2, 3, 4]:
# After left pass:  [1, 1, 2, 6]    (prefix products)
# After right pass: [24, 12, 8, 6]  (multiply by suffix products)

print("Product Except Self solution loaded ✅")
```

```python
# Solution: Best Time to Buy and Sell Stock

def max_profit_solution(prices: list[int]) -> int:
    """O(n) time, O(1) space."""
    min_price = float('inf')
    max_profit = 0
    
    for price in prices:
        min_price = min(min_price, price)
        max_profit = max(max_profit, price - min_price)
    
    return max_profit

print("Max Profit solution loaded ✅")
```

```python
# Solution: Maximum Subarray (Kadane's Algorithm)

def max_subarray_solution(nums: list[int]) -> int:
    """O(n) time, O(1) space — Kadane's Algorithm."""
    current_sum = nums[0]
    max_sum = nums[0]
    
    for i in range(1, len(nums)):
        # Either extend current subarray or start new one
        current_sum = max(nums[i], current_sum + nums[i])
        max_sum = max(max_sum, current_sum)
    
    return max_sum

# Key insight: At each position, decide:
# - Is it better to extend the previous subarray (current_sum + nums[i])?
# - Or start fresh from this element (nums[i])?

print("Maximum Subarray solution loaded ✅")
```
