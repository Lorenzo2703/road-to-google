## 📚 Theory

### The Two Pointer Pattern
The Two Pointer technique is an extremely common approach to solve array/string problems in `O(N)` time and `O(1)` space. It involves using two integer variables (pointers) that move through the array in a coordinated way.

### Common Variations
1. **Converging Pointers (Opposite Direction)**: One pointer starts at the beginning (`left = 0`), the other at the end (`right = len(arr) - 1`). They move toward the center. Commonly used when the array is **sorted** or when looking for pairs/palindromes.
2. **Same Direction (Fast and Slow)**: Both pointers start at the beginning. One moves faster than the other. Commonly used in Linked Lists (cycle detection) or when partitioning arrays in-place.

### When to Use This Pattern
- The array is **sorted** (or can be sorted, and `O(N log N)` time is acceptable).
- You need to find pairs, triplets, or sub-arrays that satisfy a specific condition.
- The naive solution requires `O(N²)` nested loops, and you need to optimize it to `O(N)`.
- You need to manipulate an array **in-place** (O(1) extra space).

### 💡 Tips for Google Interviews
- Be very careful with `while left < right:` vs `while left <= right:`. Converging pointers usually stop strictly *before* crossing, but sometimes they need to meet in the middle.
- When skipping duplicates (e.g., in `3Sum`), always check bounds: `while left < right and nums[left] == nums[left+1]`.

```python
# Solution: Valid Palindrome
def is_palindrome_solution(s: str) -> bool:
    """O(n) time, O(1) space."""
    left, right = 0, len(s) - 1
    
    while left < right:
        # Skip non-alphanumeric from left
        while left < right and not s[left].isalnum():
            left += 1
        # Skip non-alphanumeric from right
        while left < right and not s[right].isalnum():
            right -= 1
            
        if s[left].lower() != s[right].lower():
            return False
            
        left += 1
        right -= 1
        
    return True
```

```python
# Solution: Container With Most Water
def max_area_solution(height: list[int]) -> int:
    """O(n) time, O(1) space."""
    left, right = 0, len(height) - 1
    max_water = 0
    
    while left < right:
        width = right - left
        h = min(height[left], height[right])
        max_water = max(max_water, width * h)
        
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1
            
    return max_water
```

```python
# Solution: Trapping Rain Water
def trap_solution(height: list[int]) -> int:
    """O(n) time, O(1) space using two pointers."""
    if not height:
        return 0
        
    left, right = 0, len(height) - 1
    left_max, right_max = height[left], height[right]
    water = 0
    
    while left < right:
        if left_max < right_max:
            left += 1
            left_max = max(left_max, height[left])
            water += left_max - height[left]
        else:
            right -= 1
            right_max = max(right_max, height[right])
            water += right_max - height[right]
            
    return water
```
