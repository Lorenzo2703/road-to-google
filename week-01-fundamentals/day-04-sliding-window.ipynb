## 📚 Theory

### The Sliding Window Pattern
The Sliding Window is a subset of the Two Pointers pattern. It is primarily used to find a **subarray or substring** that satisfies a certain condition. Instead of restarting the search from scratch every time, a window `[left, right]` slides over the array, adding the effect of the new element and removing the effect of the old element.

### Types of Sliding Windows
1. **Fixed Window**: The size of the window `k` is constant (e.g., "find the max sum of any contiguous subarray of size k"). You move both `left` and `right` at the same time.
2. **Variable Window**: The window expands by moving `right` to include elements until a condition is met/broken, then shrinks by moving `left` to restore the condition (e.g., "find the longest substring without repeating characters").

### Variable Window Template
```python
left = 0
for right in range(len(arr)):
    # 1. Add arr[right] to the current window state
    
    while window_is_invalid():
        # 2. Remove arr[left] from the window state
        left += 1
        
    # 3. Update the global max/min answer (window is now valid)
```

### 💡 Tips for Google Interviews
- If a problem asks for **"longest/shortest subarray/substring"**, Sliding Window should immediately come to mind.
- Track the state of the window using variables like a running `sum`, a `HashSet`, or a `HashMap` (Counter).
- In hard string problems, instead of comparing two HashMaps entirely (which takes O(26)), maintain a `formed` variable that tracks how many unique characters currently meet the required frequency.

````python
# Solution: Longest Repeating Character Replacement
def character_replacement_solution(s: str, k: int) -> int:
    """O(n) time, O(1) space (at most 26 keys in the dictionary)."""
    count = {}
    left = 0
    max_freq = 0
    result = 0
    
    for right in range(len(s)):
        count[s[right]] = count.get(s[right], 0) + 1
        max_freq = max(max_freq, count[s[right]])
        
        # Current window length is (right - left + 1)
        # If the number of characters we need to replace exceeds k, shrink window
        while (right - left + 1) - max_freq > k:
            count[s[left]] -= 1
            left += 1
            
        result = max(result, right - left + 1)
        
    return result
````

````python
# Solution: Permutation in String
from collections import Counter

def check_inclusion_solution(s1: str, s2: str) -> bool:
    """O(l1 + l2) time, O(1) space (26 characters max)."""
    if len(s1) > len(s2):
        return False
        
    s1_count = Counter(s1)
    window = Counter(s2[:len(s1)])
    
    if window == s1_count:
        return True
        
    for i in range(len(s1), len(s2)):
        # Add right character
        window[s2[i]] += 1
        # Remove left character
        old_char = s2[i - len(s1)]
        window[old_char] -= 1
        
        # Important: clean up empty counts so dict comparison works cleanly
        if window[old_char] == 0:
            del window[old_char]
            
        if window == s1_count:
            return True
            
    return False
````

````python
# Solution: Minimum Window Substring
from collections import Counter

def min_window_solution(s: str, t: str) -> str:
    """O(m + n) time, O(1) space (26 chars max)."""
    if not t or not s:
        return ''
        
    t_count = Counter(t)
    required = len(t_count)
    
    window = {}
    formed = 0
    left = 0
    ans = (float('inf'), 0, 0)  # (length, left, right)
    
    for right in range(len(s)):
        c = s[right]
        window[c] = window.get(c, 0) + 1
        
        if c in t_count and window[c] == t_count[c]:
            formed += 1
            
        # The window is completely valid! Try to shrink it.
        while left <= right and formed == required:
            c_left = s[left]
            
            # Update answer if this window is smaller
            if right - left + 1 < ans[0]:
                ans = (right - left + 1, left, right)
                
            # Shrink
            window[c_left] -= 1
            if c_left in t_count and window[c_left] < t_count[c_left]:
                formed -= 1
                
            left += 1
            
    return '' if ans[0] == float('inf') else s[ans[1]:ans[2]+1]
````
