## 📚 Theory

### Hash Tables

| Operation | Average | Worst Case |
|-----------|---------|------------|
| Insert | O(1) | O(n) |
| Delete | O(1) | O(n) |
| Search | O(1) | O(n) |

### When to Use HashMaps
- **Counting frequency** of elements
- **Grouping** elements by a key
- **Two-pass lookups**: store data in pass 1, query in pass 2
- **Caching** computed results (memoization)

### Python Dict & Set Tips
- `dict.get(key, default)` — avoid KeyError
- `collections.Counter` — frequency counting
- `collections.defaultdict` — auto-initialize values
- `set()` — O(1) membership testing
- Frozensets can be used as dict keys (sets cannot)

```python
# Problem 1: Contains Duplicate (Easy) — LC #217
# Return True if any value appears at least twice

def contains_duplicate(nums: list[int]) -> bool:
    # YOUR SOLUTION HERE
    pass

# Solution:
def contains_duplicate_sol(nums: list[int]) -> bool:
    return len(nums) != len(set(nums))

assert contains_duplicate_sol([1,2,3,1]) == True
assert contains_duplicate_sol([1,2,3,4]) == False
print("✅ Contains Duplicate")
```

```python
# Problem 2: Group Anagrams (Medium) — LC #49
# Group strings that are anagrams of each other

from collections import defaultdict

def group_anagrams(strs: list[str]) -> list[list[str]]:
    # YOUR SOLUTION HERE
    pass

# Solution:
def group_anagrams_sol(strs: list[str]) -> list[list[str]]:
    groups = defaultdict(list)
    for s in strs:
        key = tuple(sorted(s))  # Anagrams have same sorted form
        groups[key].append(s)
    return list(groups.values())

# Alternative O(n*k) using character count as key:
def group_anagrams_optimal(strs: list[str]) -> list[list[str]]:
    groups = defaultdict(list)
    for s in strs:
        count = [0] * 26
        for c in s:
            count[ord(c) - ord('a')] += 1
        groups[tuple(count)].append(s)
    return list(groups.values())

print("✅ Group Anagrams")
```

```python
# Problem 3: Top K Frequent Elements (Medium) — LC #347
# Return the k most frequent elements

from collections import Counter

def top_k_frequent(nums: list[int], k: int) -> list[int]:
    # YOUR SOLUTION HERE
    pass

# Solution: Bucket Sort approach — O(n)
def top_k_frequent_sol(nums: list[int], k: int) -> list[int]:
    count = Counter(nums)
    # Bucket sort: index = frequency, value = list of nums with that freq
    buckets = [[] for _ in range(len(nums) + 1)]
    for num, freq in count.items():
        buckets[freq].append(num)
    
    result = []
    for i in range(len(buckets) - 1, -1, -1):
        for num in buckets[i]:
            result.append(num)
            if len(result) == k:
                return result
    return result

assert set(top_k_frequent_sol([1,1,1,2,2,3], 2)) == {1, 2}
print("✅ Top K Frequent Elements")
```

```python
# Problem 4: Valid Sudoku (Medium) — LC #36

def is_valid_sudoku(board: list[list[str]]) -> bool:
    # YOUR SOLUTION HERE
    pass

# Solution: Use sets for each row, column, and 3x3 box
def is_valid_sudoku_sol(board: list[list[str]]) -> bool:
    rows = [set() for _ in range(9)]
    cols = [set() for _ in range(9)]
    boxes = [set() for _ in range(9)]
    
    for r in range(9):
        for c in range(9):
            val = board[r][c]
            if val == '.':
                continue
            box_idx = (r // 3) * 3 + (c // 3)
            if val in rows[r] or val in cols[c] or val in boxes[box_idx]:
                return False
            rows[r].add(val)
            cols[c].add(val)
            boxes[box_idx].add(val)
    return True

print("✅ Valid Sudoku")
```

```python
# Problem 5: Longest Consecutive Sequence (Medium) — LC #128
# Find longest consecutive sequence in O(n) time

def longest_consecutive(nums: list[int]) -> int:
    # YOUR SOLUTION HERE
    pass

# Solution: HashSet — only start counting from sequence starts
def longest_consecutive_sol(nums: list[int]) -> int:
    num_set = set(nums)
    longest = 0
    
    for num in num_set:
        # Only start counting if num is the START of a sequence
        if num - 1 not in num_set:
            length = 1
            while num + length in num_set:
                length += 1
            longest = max(longest, length)
    
    return longest

assert longest_consecutive_sol([100, 4, 200, 1, 3, 2]) == 4
assert longest_consecutive_sol([0, 3, 7, 2, 5, 8, 4, 6, 0, 1]) == 9
print("✅ Longest Consecutive Sequence")
```
