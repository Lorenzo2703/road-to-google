## 📚 Theory

### Hash Tables
A Hash Table (or HashMap, Dictionary in Python) is a data structure that implements an associative array abstract data type, a structure that can map keys to values. A hash table uses a hash function to compute an index into an array of buckets or slots, from which the desired value can be found.

| Operation | Average Time | Worst Case | Notes |
|-----------|--------------|------------|-------|
| Insert | O(1) | O(n) | Hash collision requiring resize or traversal |
| Delete | O(1) | O(n) | Worst case if all elements hash to same bucket |
| Search | O(1) | O(n) | Fast lookups are why we use them! |

### When to Use HashMaps
- **Counting frequency** of elements.
- **Grouping** elements by a common characteristic (the key).
- **Two-pass lookups**: store data in pass 1, query in pass 2 (like Two Sum).
- **Caching** computed results (memoization).

### 🐍 Python Dict & Set Tips
- `dict.get(key, default)` — safely avoid `KeyError`.
- `collections.Counter` — extremely optimized frequency counting.
- `collections.defaultdict` — auto-initialize values (e.g., `defaultdict(list)`).
- `set()` — O(1) membership testing (like a dict with keys but no values).
- `tuple` / `frozenset` — can be used as dict keys (mutable types like `list` or `set` cannot).

### 💡 Tips for Google Interviews
- Space complexity is generally `O(N)` for HashMaps. Mention this tradeoff when proposing a HashMap solution.
- If the domain of keys is small (e.g., 26 lowercase English letters), an array of fixed size `count = [0] * 26` is more efficient than a HashMap and uses O(1) space.

```python
# Solution: Top K Frequent Elements
from collections import Counter

def top_k_frequent_solution(nums: list[int], k: int) -> list[int]:
    """O(n) time and space using Bucket Sort."""
    count = Counter(nums)
    # Index represents frequency, value is list of numbers with that frequency
    freq = [[] for _ in range(len(nums) + 1)]
    
    for num, c in count.items():
        freq[c].append(num)
        
    res = []
    # Go from right to left (highest frequency to lowest)
    for i in range(len(freq) - 1, 0, -1):
        for num in freq[i]:
            res.append(num)
            if len(res) == k:
                return res
    return res
```

```python
# Solution: Valid Sudoku
from collections import defaultdict

def is_valid_sudoku_solution(board: list[list[str]]) -> bool:
    """O(1) time and space (board is strictly 9x9)."""
    rows = defaultdict(set)
    cols = defaultdict(set)
    boxes = defaultdict(set)  # key: (r // 3, c // 3)
    
    for r in range(9):
        for c in range(9):
            val = board[r][c]
            if val == ".":
                continue
                
            if (val in rows[r] or 
                val in cols[c] or 
                val in boxes[(r // 3, c // 3)]):
                return False
                
            rows[r].add(val)
            cols[c].add(val)
            boxes[(r // 3, c // 3)].add(val)
            
    return True
```

```python
# Solution: Longest Consecutive Sequence
def longest_consecutive_solution(nums: list[int]) -> int:
    """O(n) time, O(n) space."""
    num_set = set(nums)
    longest = 0
    
    for num in num_set:
        # Only start counting if num is the START of a sequence
        if (num - 1) not in num_set:
            length = 1
            while (num + length) in num_set:
                length += 1
            longest = max(longest, length)
            
    return longest
```
