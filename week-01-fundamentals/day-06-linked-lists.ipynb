## 📚 Theory

### Linked List Basics
A Linked List is a linear data structure where elements are not stored at contiguous memory locations. Instead, elements point to the next element. The first node is called the **head**, and the last node points to `None`.

| Operation | Array | Linked List |
|-----------|-------|-------------|
| Access (by index) | O(1) | O(n) |
| Insert/Delete at Head | O(n) | O(1) |
| Insert/Delete at Tail | O(1) | O(1) (if tail pointer exists) |
| Search | O(n) | O(n) |

### Key Patterns
1. **Fast/Slow Pointers (Floyd's Tortoise and Hare)**: Used heavily for cycle detection or finding the middle of a linked list.
2. **In-place Reversal**: Changing the `next` pointers to point backwards without allocating new nodes.
3. **Dummy Head**: Using an artificial node at the beginning (`dummy = ListNode(0, head)`) to handle edge cases where the head of the list might change or be deleted.
4. **Multiple Pointers for Tracking**: In many operations (like Reorder), you need 3 pointers: `prev`, `curr`, and `nxt` to avoid losing the rest of the list when modifying `curr.next`.

### 💡 Tips for Google Interviews
- Always clarify if the list is singly or doubly linked.
- Watch out for `NullPointerException` (or `AttributeError` in Python) when doing `node.next.next` without checking if `node.next` exists.
- The dummy head pattern is an absolute must-know. It drastically simplifies insertion/deletion logic.

```python
# 🔧 Helper Code: ListNode definition and utilities
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def to_list(head: ListNode) -> list[int]:
    """Converts a linked list to a Python list for easy assertion."""
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result

def from_list(arr: list[int]) -> ListNode:
    """Converts a Python list to a linked list."""
    dummy = ListNode(0)
    curr = dummy
    for val in arr:
        curr.next = ListNode(val)
        curr = curr.next
    return dummy.next

print('ListNode helpers loaded ✅')
```

```python
# Solution: Linked List Cycle
def has_cycle_solution(head: ListNode) -> bool:
    """O(n) time, O(1) space."""
    slow = head
    fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
        if slow == fast:
            return True
            
    return False
```

```python
# Solution: Reorder List
def reorder_list_solution(head: ListNode) -> None:
    """O(n) time, O(1) space."""
    if not head or not head.next:
        return
        
    # 1. Find the middle using Fast/Slow pointers
    slow, fast = head, head.next
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
    # 2. Reverse the second half of the list
    second = slow.next
    slow.next = None  # Sever the first half from the second
    prev = None
    
    while second:
        nxt = second.next
        second.next = prev
        prev = second
        second = nxt
        
    # 3. Merge the two halves
    first, second = head, prev
    while second:
        # Save next pointers
        tmp1, tmp2 = first.next, second.next
        
        # Link the nodes
        first.next = second
        second.next = tmp1
        
        # Advance pointers
        first, second = tmp1, tmp2
```

**Why it's relevant**: Tests the "delayed pointer" or "fixed gap" technique.

**Problem**: Given the `head` of a linked list, remove the `nth` node from the end of the list and return its head.

**Approach**:

2. Initialize a `left` pointer at the dummy head, and a `right` pointer at `head`.

5. Update `left.next` to skip the target node.

```python
# Solution: Remove Nth Node From End
def remove_nth_from_end_solution(head: ListNode, n: int) -> ListNode:
    """O(n) time, O(1) space using a delayed pointer."""
    dummy = ListNode(0, head)
    left = dummy
    right = head
    
    # Advance right by n steps to establish the gap
    for _ in range(n):
        right = right.next
        
    # Slide the window to the end of the list
    while right:
        left = left.next
        right = right.next
        
    # Delete the node
    left.next = left.next.next
    
    return dummy.next
```
