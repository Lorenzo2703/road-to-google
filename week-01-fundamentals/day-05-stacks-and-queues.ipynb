## 📚 Theory

### Stack (LIFO)
A Stack follows the **Last-In, First-Out** principle. In Python, you can just use a `list` as a stack.
- **Push**: `stack.append(val)` — O(1)
- **Pop**: `stack.pop()` — O(1)
- **Peek/Top**: `stack[-1]` — O(1)
- **Use cases**: Bracket matching, undo/redo mechanisms, Depth-First Search (DFS), expression evaluation.

#### The Monotonic Stack Pattern
A Monotonic Stack is a stack whose elements are strictly increasing or strictly decreasing. It is the optimal solution for "next greater element" or "next smaller element" problems. When a new element breaks the monotonic property, you `pop()` from the stack until the property is restored.

### Queue (FIFO)
A Queue follows the **First-In, First-Out** principle. In Python, **never** use a `list` for a queue because `list.pop(0)` is O(N). Instead, use `collections.deque`.
- **Enqueue**: `queue.append(val)` — O(1)
- **Dequeue**: `queue.popleft()` — O(1)
- **Use cases**: Breadth-First Search (BFS), task scheduling, buffering.

### 💡 Tips for Google Interviews
- When dealing with intervals, histograms, or running minimums, think about maintaining state inside a stack.
- If an interviewer asks you to build a system that processes streams of data chronologically, use a Queue.

```python
# Solution: Valid Parentheses
def is_valid_solution(s: str) -> bool:
    """O(n) time, O(n) space."""
    stack = []
    pairs = {')': '(', '}': '{', ']': '['}
    
    for char in s:
        if char in pairs:
            # If stack is empty or top doesn't match
            if not stack or stack[-1] != pairs[char]:
                return False
            stack.pop()
        else:
            stack.append(char)
            
    return len(stack) == 0
```

**Approach**:
Maintain two stacks: one for the actual values, and another for the minimum values seen so far up to that level in the stack.

```python
class MinStack:
    def __init__(self):
        pass
    def push(self, val: int) -> None:
        pass
    def pop(self) -> None:
        pass
    def top(self) -> int:
        pass
    def getMin(self) -> int:
        pass

# Test cases
ms = MinStack()
ms.push(-2)
ms.push(0)
ms.push(-3)
assert ms.getMin() == -3
ms.pop()
assert ms.top() == 0
assert ms.getMin() == -2
print("All tests passed! ✅")
```

```python
# Solution: Min Stack
class MinStackSolution:
    """O(1) time for all operations, O(n) space."""
    def __init__(self):
        self.stack = []
        self.min_stack = []  # Tracks minimum at each level
    
    def push(self, val: int) -> None:
        self.stack.append(val)
        # The minimum is either the new value, or the previous minimum
        min_val = min(val, self.min_stack[-1] if self.min_stack else val)
        self.min_stack.append(min_val)
    
    def pop(self) -> None:
        self.stack.pop()
        self.min_stack.pop()
    
    def top(self) -> int:
        return self.stack[-1]
    
    def getMin(self) -> int:
        return self.min_stack[-1]
```

```python
# Solution: Daily Temperatures
def daily_temperatures_solution(temperatures: list[int]) -> list[int]:
    """O(n) time, O(n) space using Monotonic Decreasing Stack."""
    n = len(temperatures)
    result = [0] * n
    stack = []  # Stack stores indices
    
    for curr_idx, curr_temp in enumerate(temperatures):
        # While stack is not empty and current temp > temp at stack top index
        while stack and curr_temp > temperatures[stack[-1]]:
            prev_idx = stack.pop()
            result[prev_idx] = curr_idx - prev_idx
        stack.append(curr_idx)
        
    return result
```

```python
# Solution: Evaluate Reverse Polish Notation
def eval_rpn_solution(tokens: list[str]) -> int:
    """O(n) time, O(n) space."""
    stack = []
    ops = {'+', '-', '*', '/'}
    
    for token in tokens:
        if token in ops:
            b, a = stack.pop(), stack.pop()  # Note: b is popped first!
            if token == '+': stack.append(a + b)
            elif token == '-': stack.append(a - b)
            elif token == '*': stack.append(a * b)
            else: stack.append(int(a / b))  # Truncate toward zero natively in Python 3
        else:
            stack.append(int(token))
            
    return stack[0]
```

```python
# Solution: Largest Rectangle in Histogram
def largest_rectangle_area_solution(heights: list[int]) -> int:
    """O(n) time, O(n) space."""
    max_area = 0
    stack = []  # pairs: (index, height)
    
    for i, h in enumerate(heights):
        start = i
        while stack and stack[-1][1] > h:
            index, height = stack.pop()
            max_area = max(max_area, height * (i - index))
            start = index  # The current height can extend backwards to this index
        stack.append((start, h))
        
    # Calculate area for any remaining heights in the stack
    for i, h in stack:
        max_area = max(max_area, h * (len(heights) - i))
        
    return max_area
```

```python
# Solution: Implement Queue using Stacks
class MyQueueSolution:
    """Amortized O(1) time per operation, O(n) space."""
    def __init__(self):
        self.s1 = []  # For enqueueing
        self.s2 = []  # For dequeueing/peeking

    def push(self, x: int) -> None:
        self.s1.append(x)

    def pop(self) -> int:
        self.peek()  # Ensure s2 is populated
        return self.s2.pop()

    def peek(self) -> int:
        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())
        return self.s2[-1]

    def empty(self) -> bool:
        return not self.s1 and not self.s2
```
