## 📚 Theory

### Stack (LIFO)
- Push/Pop: O(1)
- Use cases: Bracket matching, undo/redo, DFS, expression evaluation
- **Monotonic Stack**: Stack where elements are in sorted order — used for 'next greater/smaller element' problems

### Queue (FIFO)
- Enqueue/Dequeue: O(1)
- Use cases: BFS, task scheduling, buffering
- **Deque**: Double-ended queue — push/pop from both ends in O(1)

```python
# Problem 1: Valid Parentheses (Easy) — LC #20

def is_valid(s: str) -> bool:
    stack = []
    pairs = {')': '(', '}': '{', ']': '['}
    for c in s:
        if c in pairs:
            if not stack or stack[-1] != pairs[c]:
                return False
            stack.pop()
        else:
            stack.append(c)
    return len(stack) == 0

assert is_valid('()[]{}') == True
assert is_valid('(]') == False
assert is_valid('([)]') == False
print('✅ Valid Parentheses')
```

```python
# Problem 2: Min Stack (Medium) — LC #155

class MinStack:
    def __init__(self):
        self.stack = []
        self.min_stack = []  # Tracks minimum at each level
    
    def push(self, val: int) -> None:
        self.stack.append(val)
        min_val = min(val, self.min_stack[-1] if self.min_stack else val)
        self.min_stack.append(min_val)
    
    def pop(self) -> None:
        self.stack.pop()
        self.min_stack.pop()
    
    def top(self) -> int:
        return self.stack[-1]
    
    def getMin(self) -> int:
        return self.min_stack[-1]

ms = MinStack()
ms.push(-2); ms.push(0); ms.push(-3)
assert ms.getMin() == -3
ms.pop()
assert ms.top() == 0
assert ms.getMin() == -2
print('✅ Min Stack')
```

```python
# Problem 3: Daily Temperatures (Medium) — LC #739
# Monotonic Stack — find next warmer day

def daily_temperatures(temperatures: list[int]) -> list[int]:
    n = len(temperatures)
    result = [0] * n
    stack = []  # Stack of indices (monotonic decreasing temperatures)
    
    for i in range(n):
        while stack and temperatures[i] > temperatures[stack[-1]]:
            prev = stack.pop()
            result[prev] = i - prev
        stack.append(i)
    
    return result

assert daily_temperatures([73,74,75,71,69,72,76,73]) == [1,1,4,2,1,1,0,0]
print('✅ Daily Temperatures')
```

```python
# Problem 4: Evaluate Reverse Polish Notation (Medium) — LC #150

def eval_rpn(tokens: list[str]) -> int:
    stack = []
    ops = {'+', '-', '*', '/'}
    for token in tokens:
        if token in ops:
            b, a = stack.pop(), stack.pop()
            if token == '+': stack.append(a + b)
            elif token == '-': stack.append(a - b)
            elif token == '*': stack.append(a * b)
            else: stack.append(int(a / b))  # Truncate toward zero
        else:
            stack.append(int(token))
    return stack[0]

assert eval_rpn(['2','1','+','3','*']) == 9
assert eval_rpn(['4','13','5','/','+']) == 6
print('✅ Evaluate RPN')
```

```python
# Problem 5: Largest Rectangle in Histogram (Hard) — LC #84

def largest_rectangle_area(heights: list[int]) -> int:
    stack = []  # Stack of indices
    max_area = 0
    
    for i, h in enumerate(heights):
        start = i
        while stack and stack[-1][1] > h:
            idx, height = stack.pop()
            max_area = max(max_area, height * (i - idx))
            start = idx
        stack.append((start, h))
    
    for idx, height in stack:
        max_area = max(max_area, height * (len(heights) - idx))
    
    return max_area

assert largest_rectangle_area([2,1,5,6,2,3]) == 10
print('✅ Largest Rectangle in Histogram')
```
