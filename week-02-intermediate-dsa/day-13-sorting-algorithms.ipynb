## 📚 Theory

While you rarely have to write a raw `MergeSort` or `QuickSort` from scratch in a Google interview, you *must* know how to leverage sorting to solve problems efficiently. Often, sorting an array first can reduce an `O(N^2)` brute-force solution down to `O(N log N)`.

### Core Sorting Concepts
1. **Comparison Sorting**: Python's `list.sort()` uses Timsort (a hybrid of Merge Sort and Insertion Sort). It runs in `O(N log N)` time and `O(N)` space. It is **stable** (preserves the relative order of equal elements).
2. **Custom Comparators**: Sometimes you need to sort by specific rules (e.g., sorting strings by length, or sorting intervals by their start times). In Python, you use the `key` argument: `intervals.sort(key=lambda x: x[0])`.
3. **In-place Sorting**: Sometimes memory is strictly limited, and you must sort the array in `O(1)` extra space. The Dutch National Flag algorithm is a prime example of an `O(N)` time, `O(1)` space specialized sort.

### Interval Problems
Intervals are heavily tested at Google. The general template is:
1. Sort the intervals by their start time.
2. Iterate through the intervals, maintaining the "current active interval".
3. If the next interval overlaps, merge them. If not, push the current interval to results and update the current active interval.

```python
# Solution: Merge Intervals
def merge_intervals_solution(intervals: list[list[int]]) -> list[list[int]]:
    """O(N log N) time, O(N) space."""
    if not intervals:
        return []
        
    intervals.sort(key=lambda x: x[0])
    merged = [intervals[0]]
    
    for i in range(1, len(intervals)):
        last = merged[-1]
        current = intervals[i]
        
        if current[0] <= last[1]:
            # Overlap! Extend the last interval's end time
            last[1] = max(last[1], current[1])
        else:
            # No overlap, add to results
            merged.append(current)
            
    return merged
```

```python
# Solution: Largest Number
def largest_number_solution(nums: list[int]) -> str:
    """O(N log N * L) time where L is max str len, O(N) space."""
    from functools import cmp_to_key
    
    # Custom comparator: return -1 if a should come before b
    def compare(n1, n2):
        if n1 + n2 > n2 + n1:
            return -1
        elif n1 + n2 < n2 + n1:
            return 1
        else:
            return 0
            
    strs = [str(num) for num in nums]
    strs.sort(key=cmp_to_key(compare))
    
    # Edge case: [0, 0] should return "0", not "00"
    res = "".join(strs)
    return "0" if res[0] == "0" else res
```

```python
# Solution: Non-overlapping Intervals
def erase_overlap_intervals_solution(intervals: list[list[int]]) -> int:
    """O(N log N) time, O(1) space."""
    intervals.sort(key=lambda x: x[0])
    removals = 0
    prev_end = intervals[0][1]
    
    for i in range(1, len(intervals)):
        start, end = intervals[i]
        
        if start >= prev_end:
            # No overlap, just update prev_end
            prev_end = end
        else:
            # Overlap! We must "remove" one interval.
            # To minimize future overlaps, we keep the interval that ends FIRST.
            # (Therefore, the active prev_end becomes the MIN of the two ends)
            removals += 1
            prev_end = min(prev_end, end)
            
    return removals
```
