## 📚 Theory

Now that we know basic recursive traversals, it's time to tackle more complex tree patterns.

### Breadth-First Search (BFS)
While DFS uses a Stack (or the call stack via recursion), BFS uses a **Queue**. By popping from the front of the queue and pushing children to the back, we naturally explore the tree level-by-level. This is essential when a problem asks for the *shortest path* or asks you to group nodes by their depth.

### Binary Search Tree (BST) Properties
A BST has a strict rule: For any node `N`, all values in its **left subtree** must be strictly less than `N.val`, and all values in its **right subtree** must be strictly greater than `N.val`. 
*(Crucial: It's not just the immediate children that must follow this rule, but the entire subtrees!)*

### LCA (Lowest Common Ancestor)
Finding the lowest node in `T` that has both `p` and `q` as descendants. This pattern relies heavily on propagating boolean/node states UP the recursive call stack.

### Tree Construction & Serialization
Trees are just objects in memory. To store them on disk or send them over a network, we must "serialize" them into a string. The easiest way is using a Pre-order traversal, putting a marker (like `"N"`) whenever we hit a `None` node.

```python
# 🔧 Helper Code: TreeNode definition
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def build_tree(values: list) -> TreeNode:
    if not values: return None
    root = TreeNode(values[0])
    queue = [root]
    i = 1
    while queue and i < len(values):
        node = queue.pop(0)
        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    return root

print('TreeNode helpers loaded ✅')
```

```python
# Solution: Construct Binary Tree
def build_tree_from_traversals_solution(preorder: list[int], inorder: list[int]) -> TreeNode:
    """O(n) time, O(n) space (using a hashmap to optimize index lookup)."""
    # Create a hashmap to instantly find the index of a value in the inorder array
    inorder_map = {val: idx for idx, val in enumerate(inorder)}
    preorder_idx = 0
    
    def build(in_left, in_right):
        nonlocal preorder_idx
        if in_left > in_right:
            return None
            
        # The current root is the current element in preorder
        root_val = preorder[preorder_idx]
        root = TreeNode(root_val)
        preorder_idx += 1
        
        # Find where this root is in the inorder array
        in_idx = inorder_map[root_val]
        
        # Build left and right subtrees
        root.left = build(in_left, in_idx - 1)
        root.right = build(in_idx + 1, in_right)
        
        return root
        
    return build(0, len(inorder) - 1)
```
