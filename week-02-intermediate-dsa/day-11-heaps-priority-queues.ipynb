## 📚 Theory

### Heaps (Priority Queues)
A Heap is a specialized tree-based data structure that satisfies the **heap property**:
- **Min-Heap**: The parent is always *less than or equal to* its children. The root is the absolute minimum.
- **Max-Heap**: The parent is always *greater than or equal to* its children. The root is the absolute maximum.

### Time Complexity
| Operation | Time Complexity |
|-----------|-----------------|
| Find Min/Max (Peek) | O(1) |
| Insert (Push) | O(log n) |
| Delete Min/Max (Pop) | O(log n) |
| Heapify (array -> heap) | O(n) |

### Python's `heapq` Module
In Python, `heapq` is built-in but it only supports **Min-Heaps** by default. To use a Max-Heap, you must multiply all values by `-1` before pushing them, and multiply by `-1` again when popping.
```python
import heapq
heap = [3, 1, 4, 1, 5]
heapq.heapify(heap)        # O(n) in-place transformation
heapq.heappush(heap, 2)    # O(log n)
min_val = heapq.heappop(heap) # O(log n)
```

### 💡 Tips for Google Interviews
- Whenever a problem asks for the **"Top K"**, **"Kth largest/smallest"**, or **"closest"** elements, immediately think of Heaps.
- To find the Kth *largest* element, maintain a **Min-Heap** of size K. When the heap size exceeds K, pop the minimum. The remaining K elements are the largest seen so far, and the root is the Kth largest!

````python
# Solution: Last Stone Weight
def last_stone_weight_solution(stones: list[int]) -> int:
    """O(n log n) time, O(n) space."""
    # Python has no max-heap, so we multiply by -1
    max_heap = [-s for s in stones]
    heapq.heapify(max_heap)
    
    while len(max_heap) > 1:
        first = heapq.heappop(max_heap)  # Most negative = largest
        second = heapq.heappop(max_heap)
        
        if first != second:
            # Push the difference back. (first is more negative than second)
            heapq.heappush(max_heap, first - second)
            
    return -max_heap[0] if max_heap else 0
````

````python
# Solution: Top K Frequent Elements
def top_k_frequent_solution(nums: list[int], k: int) -> list[int]:
    """O(n log k) time, O(n) space."""
    counts = collections.Counter(nums)
    min_heap = []
    
    for num, freq in counts.items():
        heapq.heappush(min_heap, (freq, num))
        if len(min_heap) > k:
            heapq.heappop(min_heap)
            
    # Return just the numbers
    return [num for freq, num in min_heap]
````

````python
# Solution: Find Median from Data Stream
class MedianFinderSolution:
    """O(log n) for addNum, O(1) for findMedian."""
    def __init__(self):
        # Max-heap for the smaller half
        self.small = []
        # Min-heap for the larger half
        self.large = []

    def addNum(self, num: int) -> None:
        # Push to small (max-heap, negate value)
        heapq.heappush(self.small, -num)
        
        # Make sure every element in small is <= every element in large
        if self.small and self.large and (-self.small[0] > self.large[0]):
            val = -heapq.heappop(self.small)
            heapq.heappush(self.large, val)
            
        # Handle uneven sizes (small can be at most 1 larger than large)
        if len(self.small) > len(self.large) + 1:
            val = -heapq.heappop(self.small)
            heapq.heappush(self.large, val)
        elif len(self.large) > len(self.small):
            val = heapq.heappop(self.large)
            heapq.heappush(self.small, -val)

    def findMedian(self) -> float:
        if len(self.small) > len(self.large):
            return float(-self.small[0])
        
        # If even length, median is average of two middle elements
        return (-self.small[0] + self.large[0]) / 2.0
````
