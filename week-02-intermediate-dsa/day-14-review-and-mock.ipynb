```python
# Solution: Search a 2D Matrix
def search_matrix_solution(matrix: list[list[int]], target: int) -> bool:
    """O(log(m*n)) time, O(1) space."""
    if not matrix or not matrix[0]:
        return False
        
    ROWS, COLS = len(matrix), len(matrix[0])
    left, right = 0, (ROWS * COLS) - 1
    
    while left <= right:
        mid = left + (right - left) // 2
        # Convert 1D mid index to 2D row/col
        r = mid // COLS
        c = mid % COLS
        
        if matrix[r][c] == target:
            return True
        elif matrix[r][c] < target:
            left = mid + 1
        else:
            right = mid - 1
            
    return False
```

```python
# Solution: Binary Tree Right Side View
def right_side_view_solution(root: TreeNode) -> list[int]:
    """O(n) time, O(n) space."""
    if not root:
        return []
        
    result = []
    queue = collections.deque([root])
    
    while queue:
        level_length = len(queue)
        
        for i in range(level_length):
            node = queue.popleft()
            
            # If it's the very last node in the queue for this level, we can see it!
            if i == level_length - 1:
                result.append(node.val)
                
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
                
    return result
```

```python
# Solution: Task Scheduler
def least_interval_solution(tasks: list[str], n: int) -> int:
    """O(N * log(26)) -> O(N) time, O(26) -> O(1) space."""
    # Count frequencies
    counts = collections.Counter(tasks)
    
    # Max Heap of frequencies (multiply by -1 for Python)
    max_heap = [-cnt for cnt in counts.values()]
    heapq.heapify(max_heap)
    
    time = 0
    # Queue to hold tasks that are cooling down: pairs of [cnt, idle_until_time]
    queue = collections.deque()
    
    while max_heap or queue:
        time += 1
        
        if max_heap:
            # Pop the most frequent task and "process" it
            cnt = heapq.heappop(max_heap) + 1  # +1 because it's negative
            
            # If it still has occurrences left, put it in the cooling queue
            if cnt < 0:
                queue.append([cnt, time + n])
                
        # Check if the task at the front of the queue is done cooling
        if queue and queue[0][1] == time:
            cooled_task_cnt, _ = queue.popleft()
            heapq.heappush(max_heap, cooled_task_cnt)
            
    return time
```
