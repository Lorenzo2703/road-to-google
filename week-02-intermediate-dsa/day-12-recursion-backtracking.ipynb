## 📚 Theory

Backtracking is an algorithmic paradigm that systematically searches for a solution to a problem among all available options. It does this by assuming that the solutions are represented by vectors `(v1, v2, ..., vi)` of values and by traversing, in a depth-first manner, the domains of the vectors until the solutions are found.

### The Standard Backtracking Template
Almost every backtracking problem can be solved by filling in the blanks of this template:
```python
def backtrack(start_index, current_path):
    # 1. Base case / Goal condition
    if is_complete(current_path):
        results.append(list(current_path))  # IMPORTANT: Append a copy!
        return
        
    # 2. Iterate through all possible choices from the current state
    for i in range(start_index, len(options)):
        if not is_valid(options[i]):
            continue
            
        # 3. Choose (Add option to path)
        current_path.append(options[i])
        
        # 4. Explore (Recurse)
        backtrack(i + 1, current_path)  # i + 1 if we can't reuse elements
        
        # 5. Un-choose (Backtrack)
        current_path.pop()
```

### 💡 Tips for Google Interviews
- **Copying state**: Remember that arrays/lists in Python are passed by reference. When appending `current_path` to `results`, you MUST append `list(current_path)` or `current_path[:]`. If you don't, your results array will just be full of empty lists at the end!
- Time complexities for backtracking are usually factorial `O(N!)` (for permutations) or exponential `O(2^N)` (for subsets).

````python
# Solution: Subsets
def subsets_solution(nums: list[int]) -> list[list[int]]:
    """O(n * 2^n) time, O(n) space."""
    result = []
    
    def backtrack(start_idx, current_path):
        # Every path is a valid subset
        result.append(list(current_path))
        
        # Iterate through remaining choices
        for i in range(start_idx, len(nums)):
            current_path.append(nums[i])      # Choose
            backtrack(i + 1, current_path)    # Explore (move to next index)
            current_path.pop()                # Un-choose
            
    backtrack(0, [])
    return result
````

````python
# Solution: Permutations
def permutations_solution(nums: list[int]) -> list[list[int]]:
    """O(n * n!) time, O(n) space."""
    result = []
    
    def backtrack(current_path):
        # Base case: We used all elements
        if len(current_path) == len(nums):
            result.append(list(current_path))
            return
            
        for i in range(len(nums)):
            if nums[i] in current_path:
                continue  # Skip used elements
                
            current_path.append(nums[i])
            backtrack(current_path)
            current_path.pop()
            
    backtrack([])
    return result
````

````python
# Solution: Combination Sum
def combination_sum_solution(candidates: list[int], target: int) -> list[list[int]]:
    """O(n^(target/min)) time."""
    result = []
    
    def backtrack(start_idx, current_path, current_sum):
        if current_sum == target:
            result.append(list(current_path))
            return
        
        if current_sum > target:
            return  # Prune the tree (we exceeded target)
            
        for i in range(start_idx, len(candidates)):
            current_path.append(candidates[i])
            
            # Note: We pass 'i' not 'i+1' because we can reuse the same element
            backtrack(i, current_path, current_sum + candidates[i])
            
            current_path.pop()
            
    backtrack(0, [], 0)
    return result
````

````python
# Solution: Word Search
def exist_solution(board: list[list[str]], word: str) -> bool:
    """O(N * 3^L) time where L is word length, O(L) space."""
    ROWS, COLS = len(board), len(board[0])
    path = set()
    
    def backtrack(r, c, i):
        if i == len(word):
            return True
            
        if (r < 0 or c < 0 or r >= ROWS or c >= COLS or 
            word[i] != board[r][c] or (r, c) in path):
            return False
            
        path.add((r, c))  # Choose
        
        # Explore all 4 directions
        res = (backtrack(r + 1, c, i + 1) or
               backtrack(r - 1, c, i + 1) or
               backtrack(r, c + 1, i + 1) or
               backtrack(r, c - 1, i + 1))
               
        path.remove((r, c))  # Un-choose
        return res
        
    for r in range(ROWS):
        for c in range(COLS):
            if backtrack(r, c, 0):
                return True
                
    return False
````

````python
# Solution: Letter Combinations of a Phone Number
def letter_combinations_solution(digits: str) -> list[str]:
    """O(4^N * N) time, O(N) space."""
    if not digits:
        return []
        
    phone_map = {
        "2": "abc", "3": "def", "4": "ghi", "5": "jkl",
        "6": "mno", "7": "pqrs", "8": "tuv", "9": "wxyz"
    }
    
    result = []
    
    def backtrack(idx, current_string):
        if len(current_string) == len(digits):
            result.append(current_string)
            return
            
        # Get the letters that the current digit maps to
        letters = phone_map[digits[idx]]
        
        for letter in letters:
            # For strings, we can just pass the concatenation directly
            # instead of append/pop because strings are immutable!
            backtrack(idx + 1, current_string + letter)
            
    backtrack(0, "")
    return result
````
