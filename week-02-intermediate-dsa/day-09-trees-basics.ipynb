## 📚 Theory

### Binary Trees
A Binary Tree is a hierarchical data structure where each node has at most two children, referred to as the **left child** and the **right child**. Trees are naturally recursive data structures, meaning almost all tree problems can (and should) be solved using recursion.

### Depth-First Search (DFS) Traversals
DFS dives as deep as possible into a branch before backtracking. There are three standard ways to traverse a tree using DFS:
1. **Pre-order** (`Root -> Left -> Right`): Useful for creating a copy of the tree.
2. **In-order** (`Left -> Root -> Right`): Yields elements in sorted order for Binary Search Trees (BSTs).
3. **Post-order** (`Left -> Right -> Root`): Useful for deleting trees or calculating sizes/heights (since you need child answers before the root can compute its answer).

### Breadth-First Search (BFS) Traversal
BFS explores the tree level by level. It uses a **Queue** (FIFO).
1. **Level-order**: Process nodes layer by layer. Excellent for finding the shortest path or closest nodes.

### 💡 Tips for Google Interviews
- Most tree problems are really just asking you: *"What traversal should I use?"* and *"What does each node need to return to its parent?"*
- Always define your base case first: `if not root: return ...`
- Time complexity for almost all standard tree problems is `O(N)` where N is the number of nodes. Space complexity is `O(H)` where H is the height of the tree (due to the recursive call stack).

```python
# 🔧 Helper Code: TreeNode definition
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# Helper to build tree from array (Level Order)
def build_tree(values: list) -> TreeNode:
    if not values:
        return None
    root = TreeNode(values[0])
    queue = [root]
    i = 1
    while queue and i < len(values):
        node = queue.pop(0)
        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            queue.append(node.left)
        i += 1
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            queue.append(node.right)
        i += 1
    return root

print('TreeNode helpers loaded ✅')
```

```python
# Solution: Diameter of Binary Tree
def diameter_of_binary_tree_solution(root: TreeNode) -> int:
    """O(n) time, O(h) space."""
    diameter = 0
    
    def get_depth(node):
        nonlocal diameter
        if not node:
            return 0
            
        left = get_depth(node.left)
        right = get_depth(node.right)
        
        # The diameter passing through THIS node is left + right
        diameter = max(diameter, left + right)
        
        # The depth we return to our parent is 1 + max(left, right)
        return 1 + max(left, right)
        
    get_depth(root)
    return diameter
```

```python
# Solution: Balanced Binary Tree
def is_balanced_solution(root: TreeNode) -> bool:
    """O(n) time, O(h) space."""
    def check_height(node):
        if not node:
            return 0
            
        left = check_height(node.left)
        if left == -1: return -1  # Propagate failure
        
        right = check_height(node.right)
        if right == -1: return -1 # Propagate failure
        
        # Check balance condition
        if abs(left - right) > 1:
            return -1
            
        return 1 + max(left, right)
        
    return check_height(root) != -1
```
