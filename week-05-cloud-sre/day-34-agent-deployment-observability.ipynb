## 📚 Theory

Deploying an agent is more complex than a standard web app because agent runs can take seconds or even minutes if they involve multiple tool calls and LLM steps.

### Deployment Architectures
1. **Synchronous (REST)**: Client sends request, waits for agent to finish. 
   - *Problem*: HTTP timeouts. Use only for very fast tasks.
2. **Asynchronous (Queue-Based)**: Client sends request, receives a `job_id`. Agent runs in the background. Client polls for status or gets a webhook callback.
3. **Streaming (WebSockets/SSE)**: Real-time updates. The user sees the agent's "thoughts" and tool results as they happen. This is the gold standard for AI UX.

### Guardrails and Safety
Agents must be constrained to prevent them from executing malicious commands or leaking private data. 
- **Output Filtering**: Checking for PII (Personally Identifiable Information) before sending to the client.
- **Tool Restrictions**: Ensuring the agent can only access specific database rows or directory paths.

### Monitoring in Production
- **Cost Tracking**: LLM tokens are expensive. Monitor usage per user.
- **Latency Monitoring**: Track how long each tool call takes.
- **Hallucination Detection**: Periodic sampling of logs for manual review.
