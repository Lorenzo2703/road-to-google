## 📚 Theory

A **Trie** (pronounced "try"), or Prefix Tree, is a tree data structure used to efficiently store and retrieve keys in a dataset of strings. There are various applications of this very efficient data structure, such as autocomplete and spellchecker.

### Structure of a Trie
A `TrieNode` typically looks like this:
```python
class TrieNode:
    def __init__(self):
        self.children = {}  # maps character -> TrieNode
        self.is_end = False # True if this node marks the end of a word
```
Inserting and searching for a string of length `L` takes `O(L)` time. This is vastly superior to searching through a list of strings (`O(N * L)`).

### Advanced Strings (Palindromes)
For palindrome problems, the "Expand from Center" pattern is generally the most optimal `O(N^2)` approach, avoiding the overhead of DP.

````python
# Solution: Implement Trie (Prefix Tree)
class TrieNodeSolution:
    def __init__(self):
        self.children = {}
        self.isWord = False

class TrieSolution:
    def __init__(self):
        self.root = TrieNodeSolution()

    def insert(self, word: str) -> None:
        curr = self.root
        for c in word:
            if c not in curr.children:
                curr.children[c] = TrieNodeSolution()
            curr = curr.children[c]
        curr.isWord = True

    def search(self, word: str) -> bool:
        curr = self.root
        for c in word:
            if c not in curr.children:
                return False
            curr = curr.children[c]
        return curr.isWord

    def startsWith(self, prefix: str) -> bool:
        curr = self.root
        for c in prefix:
            if c not in curr.children:
                return False
            curr = curr.children[c]
        return True
````

````python
# Solution: Design Add and Search Words Data Structure
class WordDictionarySolution:
    def __init__(self):
        self.root = TrieNodeSolution()

    def addWord(self, word: str) -> None:
        curr = self.root
        for c in word:
            if c not in curr.children:
                curr.children[c] = TrieNodeSolution()
            curr = curr.children[c]
        curr.isWord = True

    def search(self, word: str) -> bool:
        def dfs(j, root):
            curr = root
            
            for i in range(j, len(word)):
                c = word[i]
                if c == ".":
                    for child in curr.children.values():
                        if dfs(i + 1, child):
                            return True
                    return False
                else:
                    if c not in curr.children:
                        return False
                    curr = curr.children[c]
            return curr.isWord
            
        return dfs(0, self.root)
````

````python
# Solution: Longest Palindromic Substring
def longest_palindrome_solution(s: str) -> str:
    """O(N^2) time, O(1) space."""
    res = ""
    resLen = 0
    
    for i in range(len(s)):
        # Odd length palindromes
        l, r = i, i
        while l >= 0 and r < len(s) and s[l] == s[r]:
            if (r - l + 1) > resLen:
                res = s[l : r + 1]
                resLen = r - l + 1
            l -= 1
            r += 1
            
        # Even length palindromes
        l, r = i, i + 1
        while l >= 0 and r < len(s) and s[l] == s[r]:
            if (r - l + 1) > resLen:
                res = s[l : r + 1]
                resLen = r - l + 1
            l -= 1
            r += 1
            
    return res
````
