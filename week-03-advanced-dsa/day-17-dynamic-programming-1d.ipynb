## 📚 Theory

Dynamic Programming (DP) is simply **optimized recursion**. When a recursive algorithm solves the same subproblem multiple times, we can cache (memoize) the answer to save time, or build the answers from the bottom-up (tabulation).

### The 3 Steps to Mastering DP
1. **Recursive Backtracking**: Write a slow, brute-force recursion that tries all paths. (This is what we learned in Week 2!).
2. **Top-Down Memoization**: Add a `cache` hashmap to the recursion. If `cache[state]` exists, return it. If not, compute it, save it, and then return it.
3. **Bottom-Up Tabulation**: Create a 1D or 2D array and iteratively build up the solution from the base cases up to the final answer. This avoids the overhead of recursion entirely.

### 💡 Tips for Google Interviews
- If you get stuck on a DP problem, **always** write the recursive backtracking solution first. Once you have that, adding a cache is literally 3 extra lines of code. An `O(N)` memoized solution is just as good as a Bottom-Up DP solution in a Google interview!

```python
# Solution: Climbing Stairs
def climb_stairs_solution(n: int) -> int:
    """O(N) time, O(1) space."""
    if n <= 3:
        return n
        
    one_step_before = 2
    two_steps_before = 1
    total = 0
    
    for i in range(3, n + 1):
        total = one_step_before + two_steps_before
        two_steps_before = one_step_before
        one_step_before = total
        
    return total
```

```python
# Solution: Word Break
def word_break_solution(s: str, wordDict: list[str]) -> bool:
    """O(N * M) time where M is wordDict size, O(N) space."""
    dp = [False] * (len(s) + 1)
    dp[len(s)] = True # Base case: end of string is valid
    
    for i in range(len(s) - 1, -1, -1):
        for word in wordDict:
            if (i + len(word)) <= len(s) and s[i : i + len(word)] == word:
                dp[i] = dp[i + len(word)]
            if dp[i]:
                break
                
    return dp[0]
```

```python
# Solution: Longest Increasing Subsequence
def length_of_lis_solution(nums: list[int]) -> int:
    """O(N^2) time, O(N) space. (Note: O(N log N) is possible with Binary Search!)."""
    LIS = [1] * len(nums)
    
    # Compute backwards
    for i in range(len(nums) - 1, -1, -1):
        for j in range(i + 1, len(nums)):
            if nums[i] < nums[j]:
                LIS[i] = max(LIS[i], 1 + LIS[j])
                
    return max(LIS)
```
