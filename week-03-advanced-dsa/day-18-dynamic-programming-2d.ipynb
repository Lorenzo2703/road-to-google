## 📚 Theory

2D Dynamic Programming implies that your "state" requires two variables to be uniquely identified. Typically, this is:
1. `(row, col)` in a grid/matrix.
2. `(index1, index2)` when comparing two separate strings or arrays.
3. `(index, current_capacity)` in Knapsack-style problems.

### Memoization vs Tabulation
In 2D DP, memoization is often much faster to write because you just use `memo[(r, c)]`. Tabulation requires initializing a 2D matrix (like `dp = [[0] * COLS for _ in range(ROWS)]`) and being very careful with your loop boundaries.

### 💡 Tips for Google Interviews
- When creating a 2D array in Python, **NEVER** do `dp = [[0] * n] * m`. It creates `m` references to the *same* list. Changing one row changes all of them! Always do `[[0] * n for _ in range(m)]`.

```python
# Solution: Unique Paths
def unique_paths_solution(m: int, n: int) -> int:
    """O(M*N) time, O(N) space (optimized to 1D)."""
    # Note: We only need to store the previous row!
    row = [1] * n
    
    for i in range(m - 1):
        newRow = [1] * n
        for j in range(n - 2, -1, -1):
            newRow[j] = newRow[j + 1] + row[j]
        row = newRow
        
    return row[0]
```

```python
# Solution: Longest Common Subsequence
def longest_common_subsequence_solution(text1: str, text2: str) -> int:
    """O(M*N) time, O(M*N) space."""
    dp = [[0 for j in range(len(text2) + 1)] for i in range(len(text1) + 1)]
    
    for i in range(len(text1) - 1, -1, -1):
        for j in range(len(text2) - 1, -1, -1):
            if text1[i] == text2[j]:
                dp[i][j] = 1 + dp[i + 1][j + 1]
            else:
                dp[i][j] = max(dp[i][j + 1], dp[i + 1][j])
                
    return dp[0][0]
```

```python
# Solution: Target Sum
def find_target_sum_ways_solution(nums: list[int], target: int) -> int:
    """O(N * sum(nums)) time, O(N * sum(nums)) space."""
    dp = {} # (index, total) -> number of ways
    
    def backtrack(i, total):
        if i == len(nums):
            return 1 if total == target else 0
            
        if (i, total) in dp:
            return dp[(i, total)]
            
        dp[(i, total)] = (backtrack(i + 1, total + nums[i]) +
                          backtrack(i + 1, total - nums[i]))
                          
        return dp[(i, total)]
        
    return backtrack(0, 0)
```

```python
# Solution: Edit Distance
def min_distance_solution(word1: str, word2: str) -> int:
    """O(M*N) time, O(M*N) space."""
    dp = [[float("inf")] * (len(word2) + 1) for i in range(len(word1) + 1)]
    
    for j in range(len(word2) + 1):
        dp[len(word1)][j] = len(word2) - j
    for i in range(len(word1) + 1):
        dp[i][len(word2)] = len(word1) - i
        
    for i in range(len(word1) - 1, -1, -1):
        for j in range(len(word2) - 1, -1, -1):
            if word1[i] == word2[j]:
                dp[i][j] = dp[i + 1][j + 1]
            else:
                dp[i][j] = 1 + min(dp[i + 1][j],    # Delete
                                   dp[i][j + 1],    # Insert
                                   dp[i + 1][j + 1])# Replace
                                   
    return dp[0][0]
```
