## 📚 Theory

A greedy algorithm makes the locally optimal choice at each stage with the hope of finding a global optimum. In many problems, a greedy strategy does not produce an optimal solution, but for some specific problems (often involving max/min arrays, jumps, or intervals), it works perfectly and reduces `O(N^2)` DP down to `O(N)` time and `O(1)` space.

### Recognizing Greedy Problems
- If you are asked for the **maximum** or **minimum** of something, and making a choice *now* doesn't negatively impact future choices, it might be greedy.
- If DP gives you a "Time Limit Exceeded", look for a greedy pattern (like tracking a single max variable instead of an entire array).

### 💡 Tips for Google Interviews
- Greedy algorithms are often hard to *prove* correct during an interview. If you spot a greedy pattern, state it as a hypothesis: *"I think we can just track the maximum reach so far... let's trace it to see if it holds."*

```python
# Solution: Maximum Subarray (Kadane's Algorithm)
def max_sub_array_solution(nums: list[int]) -> int:
    """O(N) time, O(1) space."""
    max_sum = nums[0]
    current_sum = 0
    
    for num in nums:
        if current_sum < 0:
            current_sum = 0
        current_sum += num
        max_sum = max(max_sum, current_sum)
        
    return max_sum
```

```python
# Solution: Jump Game
def can_jump_solution(nums: list[int]) -> bool:
    """O(N) time, O(1) space."""
    goal = len(nums) - 1
    
    for i in range(len(nums) - 1, -1, -1):
        if i + nums[i] >= goal:
            goal = i
            
    return goal == 0
```

```python
# Solution: Partition Labels
def partition_labels_solution(s: str) -> list[int]:
    """O(N) time, O(1) space (hashmap max size is 26)."""
    # Find the last index of every character
    last_idx = {c: i for i, c in enumerate(s)}
    
    res = []
    size = 0
    end = 0
    
    for i, c in enumerate(s):
        size += 1
        # Greedily extend the end of our current partition
        end = max(end, last_idx[c])
        
        if i == end:
            res.append(size)
            size = 0
            
    return res
```
