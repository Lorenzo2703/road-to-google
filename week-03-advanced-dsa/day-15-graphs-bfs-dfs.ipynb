## 📚 Theory

Graphs are networks consisting of **vertices** (nodes) connected by **edges**. They can be directed or undirected, weighted or unweighted. In interviews, graphs are often represented as an Adjacency List (`{node: [neighbors]}`) or a 2D Grid/Matrix.

### Graph DFS
Explores as far as possible along each branch before backtracking.
- Requires a `visited` set to prevent infinite loops (cycles).
- Great for finding connected components, cycle detection, and topological sorting.

### Graph BFS
Explores the neighbor nodes first, before moving to the next level neighbors.
- Uses a `collections.deque`.
- Excellent for finding the **shortest path** on unweighted graphs.

### 💡 Tips for Google Interviews
- **Matrix as a Graph**: If you get a 2D grid of 1s and 0s, treat each cell as a node, and its up/down/left/right adjacent cells as its neighbors.
- **Cycle Detection**: In directed graphs, keeping a `path` or `visiting` set for the current DFS stack is crucial to detect cycles (used in Course Schedule problems).

```python
# Solution: Number of Islands
def num_islands_solution(grid: list[list[str]]) -> int:
    """O(M*N) time, O(M*N) space (worst case recursion depth)."""
    if not grid:
        return 0
        
    rows, cols = len(grid), len(grid[0])
    islands = 0
    
    def dfs(r, c):
        if r < 0 or c < 0 or r >= rows or c >= cols or grid[r][c] == '0':
            return
            
        # Sink the island piece
        grid[r][c] = '0'
        
        dfs(r + 1, c)
        dfs(r - 1, c)
        dfs(r, c + 1)
        dfs(r, c - 1)
        
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == '1':
                islands += 1
                dfs(r, c)
                
    return islands
```

```python
# Solution: Max Area of Island
def max_area_of_island_solution(grid: list[list[int]]) -> int:
    """O(M*N) time, O(M*N) space."""
    rows, cols = len(grid), len(grid[0])
    max_area = 0
    
    def dfs(r, c):
        if r < 0 or c < 0 or r >= rows or c >= cols or grid[r][c] == 0:
            return 0
            
        grid[r][c] = 0
        
        return (1 + dfs(r + 1, c) +
                    dfs(r - 1, c) +
                    dfs(r, c + 1) +
                    dfs(r, c - 1))
                    
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == 1:
                max_area = max(max_area, dfs(r, c))
                
    return max_area
```

**LeetCode #417** | [Link](https://leetcode.com/problems/pacific-atlantic-water-flow/)

**Why it's relevant**: Tests reverse-thinking in graphs. Sometimes it's better to start from the destination and search backward!

**Problem**: Given an `m x n` matrix of `heights`. The Pacific ocean touches the top and left edges, and the Atlantic touches the bottom and right edges. Water can only flow to a neighboring cell with an equal or lower height. Return a list of grid coordinates where water can flow to BOTH oceans.

```python
# Solution: Pacific Atlantic Water Flow
def pacific_atlantic_solution(heights: list[list[int]]) -> list[list[int]]:
    """O(M*N) time, O(M*N) space."""
    rows, cols = len(heights), len(heights[0])
    pac, atl = set(), set()
    
    def dfs(r, c, reachable_set, prev_height):
        if (r < 0 or c < 0 or r >= rows or c >= cols or 
            (r, c) in reachable_set or heights[r][c] < prev_height):
            return
            
        reachable_set.add((r, c))
        
        dfs(r + 1, c, reachable_set, heights[r][c])
        dfs(r - 1, c, reachable_set, heights[r][c])
        dfs(r, c + 1, reachable_set, heights[r][c])
        dfs(r, c - 1, reachable_set, heights[r][c])
        
    for c in range(cols):
        dfs(0, c, pac, heights[0][c])        # Pacific top row
        dfs(rows - 1, c, atl, heights[rows - 1][c]) # Atlantic bottom row
        
    for r in range(rows):
        dfs(r, 0, pac, heights[r][0])        # Pacific left col
        dfs(r, cols - 1, atl, heights[r][cols - 1]) # Atlantic right col
        
    res = []
    for r in range(rows):
        for c in range(cols):
            if (r, c) in pac and (r, c) in atl:
                res.append([r, c])
                
    return res
```

```python
# Solution: Course Schedule
def can_finish_solution(numCourses: int, prerequisites: list[list[int]]) -> bool:
    """O(V + E) time, O(V + E) space."""
    preMap = {i: [] for i in range(numCourses)}
    for crs, pre in prerequisites:
        preMap[crs].append(pre)
        
    visiting = set()
    
    def dfs(crs):
        if crs in visiting:
            return False # Cycle detected!
        if preMap[crs] == []:
            return True  # Known good course
            
        visiting.add(crs)
        
        for pre in preMap[crs]:
            if not dfs(pre):
                return False
                
        visiting.remove(crs)
        preMap[crs] = [] # Cache result so we don't re-run DFS on this node
        return True
        
    for c in range(numCourses):
        if not dfs(c):
            return False
            
    return True
```
