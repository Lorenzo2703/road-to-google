## 📚 Theory: From Code to Production

Writing scalable code is only half the battle. In a modern cloud environment, getting that code running securely and reliably across thousands of servers is an engineering challenge of its own.

### 1. Containers (Docker)
Before containers, developers would say, "Well, it works on my machine!" when code crashed in production due to missing dependencies or different OS versions.
A **Container** bundles the application code along with all its dependencies, libraries, and configuration files into a single immutable artifact. This guarantees it will run exactly the same way anywhere.

### 2. Orchestration (Kubernetes / K8s)
If your application needs to scale to 500 instances, you can't manually start 500 containers and keep track of which ones crash. 
**Kubernetes** is an open-source system (originally developed by Google as "Borg") for automating deployment, scaling, and management of containerized applications.
- **Pod**: The smallest deployable unit in K8s, usually containing one container.
- **Deployment**: Manages the declarative state of Pods (e.g., "Ensure exactly 10 replicas of this Pod are running at all times. If one dies, restart it.").
- **Service**: An abstract way to expose an application running on a set of Pods as a network service (internal load balancing).

## 🔨 Exercise: CI/CD and Deployment Strategies

Continuous Integration and Continuous Deployment (CI/CD) is the automated process of testing and releasing code.

**The CI/CD Pipeline**:
1. **Commit**: Developer pushes code to GitHub.
2. **Build**: CI server (like GitHub Actions or Jenkins) pulls the code, runs unit tests, and compiles the binaries.
3. **Package**: CI server builds a new Docker Image and pushes it to a Container Registry.
4. **Deploy**: CD tool (like ArgoCD) detects the new image and commands Kubernetes to roll it out to production.

### Deployment Strategies
When deploying version 2.0 (V2) to replace version 1.0 (V1), you have several options:

#### 1. Rolling Deployment (Standard)
Gradually replace V1 Pods with V2 Pods one by one. 
- *Pros*: Zero downtime.
- *Cons*: Both V1 and V2 are running simultaneously during the rollout, which can cause issues if they share a database schema that V2 altered.

#### 2. Blue / Green Deployment
You have two identical production environments (Blue and Green). Blue is currently live running V1. You deploy V2 entirely to the Green environment. Once Green passes all health checks, you flip the Load Balancer to point 100% of traffic to Green.
- *Pros*: Instant rollback (just flip the Load Balancer back to Blue if Green crashes). No version mixing.
- *Cons*: You have to pay for 2x the infrastructure costs!

#### 3. Canary Deployment
Named after the "canary in the coal mine." You deploy V2 to a tiny subset of your servers and route exactly 1% of live user traffic to it. You carefully monitor the error rates and metrics for V2. If it's stable, you slowly dial it up to 10%, 50%, and then 100%.
- *Pros*: Minimizes blast radius. If V2 has a catastrophic bug, it only affects 1% of users before being caught.
- *Cons*: Complex to configure load balancers and routing logic.
