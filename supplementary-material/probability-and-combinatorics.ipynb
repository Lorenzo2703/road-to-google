## 📚 Theory: Combinatorics

Combinatorics is the study of counting, grouping, and arranging objects.

### Permutations ($n!$)
The number of ways to arrange $n$ distinct objects in a sequence.
- Formula: $P(n, r) = \frac{n!}{(n-r)!}$

### Combinations (N choose K)
The number of ways to choose $k$ objects from a set of $n$ distinct objects, where the order of selection does not matter.
- Formula: $\binom{n}{k} = \frac{n!}{k!(n-k)!}$

### The Pigeonhole Principle
If $n$ items are put into $m$ containers, with $n > m$, then at least one container must contain more than one item.

## 📚 Theory: Elementary Probability

### Basic Probability
$P(A) = \frac{\text{Number of favorable outcomes}}{\text{Total number of outcomes}}$

### Expected Value
The long-run average value of repetitions of the same experiment.
- $E[X] = \sum x \cdot P(X=x)$

### Conditional Probability (Bayes' Theorem)
$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$

## 🔨 Practice Problems

**1. N Choose K Implementation**:
Implement a function to calculate $\binom{n}{k}$ efficiently without causing integer overflow too early.

```python
def n_choose_k(n, k):
    if k < 0 or k > n:
        return 0
    if k == 0 or k == n:
        return 1
    if k > n // 2:
        k = n - k
    
    numerator = 1
    for i in range(k):
        numerator = numerator * (n - i) // (i + 1)
    return numerator

assert n_choose_k(5, 2) == 10
assert n_choose_k(10, 3) == 120
print("n_choose_k passed! ✅")
```

**2. The Birthday Paradox**:
What is the probability that in a room of 23 people, at least two people share the same birthday?

**Answer**: It's easier to calculate the probability that NO two people share a birthday: 
$P(\text{no match}) = \frac{365}{365} \times \frac{364}{365} \times \dots \times \frac{365-22}{365} \approx 0.493$
Therefore, $P(\text{at least one match}) = 1 - 0.493 = 0.507$ or about 50.7%.
