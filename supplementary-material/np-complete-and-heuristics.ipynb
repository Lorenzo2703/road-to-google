## 📚 Theory: P vs NP

- **P (Polynomial Time)**: Problems that can be *solved* quickly (e.g., Sorting, Shortest Path).
- **NP (Nondeterministic Polynomial Time)**: Problems where a solution can be *verified* quickly.
- **NP-Complete**: The "hardest" problems in NP. If you find a polynomial-time solution for one, you solve all of P vs NP. (e.g., Sudoku, Traveling Salesman).

### 1. Traveling Salesman Problem (TSP)
**Problem**: Given a list of cities and the distances between each pair, what is the shortest possible route that visits each city exactly once and returns to the origin city?
- **Complexity**: $O(n!)$ brute force, $O(n^2 2^n)$ with Dynamic Programming (Held-Karp).
- **Heuristics**: In real-world systems, we use **Nearest Neighbor** (greedy) or **Ant Colony Optimization** to find a "good enough" solution quickly.

### 2. 0/1 Knapsack Problem
**Problem**: Given a set of items, each with a weight and a value, determine the number of each item to include in a collection so that the total weight is less than or equal to a given limit and the total value is as large as possible.
- **Complexity**: $O(N \times W)$ using Dynamic Programming. 
- **Note**: While $O(N \times W)$ looks polynomial, it's actually "Pseudo-polynomial" because it depends on the numerical value of $W$, not the number of bits.

## 🔨 Practice: Knapsack DP Implementation

**Task**: Implement the 0/1 Knapsack problem using Bottom-Up DP.

```python
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]
    
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i-1] <= w:
                dp[i][w] = max(values[i-1] + dp[i-1][w - weights[i-1]], 
                               dp[i-1][w])
            else:
                dp[i][w] = dp[i-1][w]
                
    return dp[n][capacity]

assert knapsack([1, 2, 3], [10, 15, 40], 6) == 65 # 15 + 40 + 10 (total weight 6)
print("Knapsack passed! ✅")
```

## 📝 Theory: Heuristics & Approximation

When a problem is NP-Complete, don't try to find the perfect answer in an interview unless $N$ is very small. Instead:
1. **State that it is NP-Complete**.
2. **Propose a Greedy approach** (e.g., always pick the city with the highest value-to-weight ratio).
3. **Discuss Local Search** (e.g., Simulated Annealing).
