# Interview Questions Deep Dive

**Strategic Interviewing & Test Design**

---

## 🧪 1. Testing & Quality Assurance

### Q: "How would you unit test the code you write?"
**Answer Strategy**: Mention the **Test Pyramid**. Unit tests at the bottom (fast, many), Integration tests in the middle, E2E tests at the top (slow, few).
- **Isolation**: Use Mocks/Stubs for external services (DB, APIs).
- **State**: Reset state between tests to ensure independence.
- **Data-Driven Testing**: Test the same logic with a wide array of input/output pairs.

### Q: "How would you design E2E, Load, and Security tests for Gmail?"
**Deep Dive**:
- **Chaos Engineering**: Google uses tools like 'Chubby' or internal fault-injection to randomly kill servers in production. Ask: "Does the system remain available?"
- **A/B Testing**: How do we test a new UI feature? Roll it out to 1% of users (Canary), measure engagement metrics (click-through rate), and compare against the control group.
- **Backpressure**: In load testing, simulate what happens when the 'Compose' service is slow. Does the API Gateway queue requests or drop them with 429 errors?

## 🧠 2. Edge Case Masterlist

When asked for "interesting test cases", use this cheat sheet categorized by Data Structure:

| Data Structure | Critical Edge Cases |
|----------------|---------------------|
| **Linked List** | Empty list, single node, cycle in the list, removing the head/tail. |
| **Trees** | Empty tree, skewed tree (looks like a linked list), single node. |
| **Graphs** | Disconnected graph (islands), cyclic graph, self-loops, weighted vs unweighted. |
| **Strings** | Empty string, single character, all spaces, Unicode/UTF-8, very long strings. |
| **Arrays** | Empty array, single element, duplicates, all elements identical, very large numbers (overflow). |
| **Recursion** | Base case not reached (stack overflow), negative input for factorial/fibonacci. |

## 📊 3. Advanced Probabilistic & Combinatorial

### Q: The Monty Hall Problem
**Scenario**: You're on a game show with 3 doors. Behind one is a car, behind others are goats. You pick Door 1. The host (who knows what's behind doors) opens Door 3 to reveal a goat. He asks: "Do you want to switch to Door 2?"
- **Answer**: **Yes, always switch**. Your original pick had a 1/3 chance of being right. The remaining 2/3 chance is now concentrated entirely on Door 2. Switching gives you a 66% win rate.

### Q: Expected Value (Coins)
**Question**: You flip a fair coin until you get 'Heads'. What is the expected number of flips?
- **Answer**: **2**. 
- **Formula**: This is a Geometric Distribution with $p=0.5$. $E[X] = 1/p = 1/0.5 = 2$.

### Q: N Choose K (Stars and Bars)
**Question**: How many ways can you distribute 10 identical apples among 3 different children?
- **Answer**: Use the "Stars and Bars" formula: $\binom{n+k-1}{k-1} = \binom{10+3-1}{3-1} = \binom{12}{2} = 66$.

## 🏗️ 4. System Design "Hidden" Questions

### Q: Clock Drift in Distributed Systems
**Problem**: Two servers in different data centers think it is 10:00:01 AM. But one is 50ms ahead. If they both write to a DB, which write "wins"?
- **Answer**: Don't rely on wall-clock time. Use **Logical Clocks** (Lamport Clocks) or **Vector Clocks** to establish causality (which event actually happened first).

### Q: Cascading Failures & Circuit Breakers
**Problem**: Service A calls Service B. Service B is slow. Service A's threads all wait for B, and eventually, Service A crashes. Then the next service calling A crashes. This is a cascading failure.
- **Answer**: Use a **Circuit Breaker** (like Hystrix). If Service B is slow, the breaker "trips" and Service A immediately returns an error/cached response without waiting, protecting Service A's resources.

## 💻 5. Coding Design Patterns

| Pattern | When to Use | Example in Agents |
|---------|-------------|--------------------|
| **Strategy** | Switching algorithms at runtime. | Switching between Gemini Pro and Gemini Flash based on task complexity. |
| **Factory** | Creating objects without specifying the exact class. | An `AgentFactory` that creates a 'CoderAgent' or 'ResearcherAgent' based on a config string. |
| **Singleton** | Ensuring a class has only one instance. | A global `DatabaseConnection` or `ConfigurationManager` instance. |
| **Observer** | One-to-many dependency notification. | Notifying multiple UI components when an Agent's status changes. |

## 📝 Final Interview Prep Checklist
- [ ] **Ask Clarifying Questions**: "Is this for a single user or millions?", "Is the data sorted?"
- [ ] **Think Out Loud**: Explain your trade-offs while you code.
- [ ] **Optimize**: After solving, ask "Can we do better than O(N^2)?"
- [ ] **Verify**: Mentally step through your code with an example.
