## 📚 Theory: Operating System Basics

Understanding how the OS manages resources is critical for building performant and thread-safe systems.

### Processes vs. Threads
- **Process**: An instance of a program in execution. It has its own address space, memory, and resources. Processes are isolated from each other.
- **Thread**: The smallest unit of execution within a process. Threads share the same address space and memory of the parent process, making communication between them faster but riskier.

### Context Switching
The process of storing the state of a process or thread so that it can be restored and resume execution at a later point. This allows multiple processes to share a single CPU.

### Concurrency Issues
- **Race Condition**: Occurs when multiple threads access and manipulate the same data concurrently, and the outcome depends on the particular order in which the access takes place.
- **Deadlock**: A state where two or more threads are blocked forever, each waiting for the other to release a resource.
- **Livelock**: Similar to deadlock, but the threads are not blocked; they are continuously changing their state in response to each other without making any progress.

### Synchronization Primitives
- **Lock/Mutex (Mutual Exclusion)**: A mechanism to ensure only one thread can access a resource at a time.
- **Semaphore**: A signaling mechanism that can allow a fixed number of threads to access a resource (a mutex is a binary semaphore).
- **Monitor**: A high-level synchronization construct that encapsulates data and the procedures to manipulate it, ensuring mutual exclusion.
- **Spinlock**: A lock where the thread "spins" in a loop, repeatedly checking if the lock is available (useful for very short wait times).

## 🔨 Practice: Identifying Deadlock Conditions

For a deadlock to occur, these four conditions must hold simultaneously (Coffman Conditions):
1. **Mutual Exclusion**: At least one resource must be held in a non-sharable mode.
2. **Hold and Wait**: A thread must be holding at least one resource and waiting to acquire additional resources held by other threads.
3. **No Preemption**: Resources cannot be taken away from a thread; they must be released voluntarily.
4. **Circular Wait**: A set of threads must exist such that every thread is waiting for a resource held by the next thread in the set.

**Exercise**: How would you prevent a circular wait in a system with multiple locks?

### Answer:
**Lock Ordering**: Impose a strict order in which resources/locks must be acquired. If all threads acquire locks in the same order (e.g., always Lock A before Lock B), a circular wait becomes impossible.
