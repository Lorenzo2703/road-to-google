## 📚 Theory

At Google, behavioral interviews aren't just "chatting". They are structured assessments of how you handle real-world challenges. For SWE II, the focus is on **ownership, collaboration, and learning from failure**.

### The STAR Method
When answering any "Tell me about a time when..." question, use this structure:

1. **Situation**: Set the scene. (1-2 sentences)
2. **Task**: What was the challenge or goal? (1-2 sentences)
3. **Action**: What did **YOU** specifically do? This is the most important part. (4-5 sentences)
4. **Result**: What was the outcome? Use numbers if possible (e.g., "reduced latency by 20%"). (1-2 sentences)

### Common Themes
- **Conflict Resolution**: How did you handle a disagreement with a peer or manager?
- **Ambiguity**: How did you proceed when requirements were unclear?
- **Failure**: Tell me about a time you made a mistake. What did you learn?

## 💡 Interview Pro-Tip

Avoid using "We". Google wants to know what **YOU** did. Even if it was a team effort, focus on your specific contribution to the solution.
