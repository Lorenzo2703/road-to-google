## 🚀 Last-Minute Interview Tips

1. **Think Aloud**: The interviewer can't help you if they don't know what you're thinking.
2. **Be Collaborative**: Treat the interviewer as a teammate, not an adversary.
3. **Stay Positive**: If you struggle with a problem, stay calm and keep working through it. Perseverance is a key trait of Googleyness.
4. **Ask Good Questions**: When asked "Do you have any questions for us?", have 2-3 prepared that show your interest in the team and the technology (e.g., "How does the team handle the non-deterministic nature of agents in production?").
