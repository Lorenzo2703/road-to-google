## 📚 Theory

In the final stages of the interview process, the hiring committee looks for **Role Fit**. For the "Agent Development" role, this means proving you aren't just a generic SWE, but someone who understands the unique challenges of building AI agents.

### Your Narrative
You need to be able to explain:
- **What** you've built with LLMs/Agents.
- **Why** you made certain architectural choices (e.g., "I used a vector database because...").
- **How** you handled the specific non-deterministic nature of AI (e.g., "I implemented a custom evaluation rubric using Gemini...").

### Questions to Master
- "Why are you interested in Agent Development?"
- "What is the most challenging problem you've solved in the AI space?"
- "How do you see the field of Autonomous Agents evolving in the next 2 years?"
