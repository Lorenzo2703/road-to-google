## 📚 Theory

In your final system design review, we focus on the "Google Scale" bottlenecks. Specifically, how to handle millions of requests while maintaining low latency and high availability.

### Key Architecture Patterns
- **Microservices vs Monolith**: Understand when to break a service apart (e.g., separating the 'Search' service from the 'Booking' service).
- **Data Sharding**: How to split your database across multiple servers (e.g., sharding by `user_id`).
- **Rate Limiting**: Protecting your internal services from being overwhelmed by traffic spikes.
- **Message Queues**: Handling long-running tasks asynchronously (crucial for LLM agents).

### Special Focus: LLM System Design
When designing an agent-based system:
- **Caching**: Use semantic caching (caching similar queries, not just exact matches).
- **Batching**: Combining multiple small LLM requests into a single larger one to reduce overhead.
- **Model Routing**: Routing simple queries to a small/fast model and complex ones to a large/powerful model.
