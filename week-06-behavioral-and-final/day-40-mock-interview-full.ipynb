# Day 40: Full Mock Interview Marathon

**Week 6 — Behavioral & Final Prep**

---

## 📚 The Schedule

Today you simulate a full Google "Onsite" (virtual) interview. Block out 4 hours. No distractions. No Google-ing for answers.

### Round 1: Coding (45 mins)
**Problem**: Design a highly efficient LRU (Least Recently Used) Cache that supports `get` and `put` in `O(1)` time.

### Round 2: System Design (45 mins)
**Problem**: Design the architecture for a large-scale **AI Agent Marketplace**. Users can upload agents, and other users can pay to run them. Consider security (running untrusted code), billing, and latency.

### Round 3: Behavioral (45 mins)
**Questions**:
1. Tell me about a time you led a project with an ambiguous goal.
2. How do you handle a team member who is not pulling their weight?
3. Tell me about a time you disagreed with a technical decision made by your manager.

### Round 4: Coding (45 mins)
**Problem**: Find the length of the longest path in a Binary Tree where each node has a value, and the path consists of nodes with the same value.

## 📋 Post-Interview Reflection

After the 4 hours, evaluate yourself on:
- **Stamina**: Did your performance drop in Round 4?
- **Communication**: Did you keep talking while you were stuck?
- **Technical Accuracy**: Did you miss edge cases (empty inputs, large scale)?
- **Googleyness**: Did you stay calm and collaborative?

## 📝 Day 40 Review

### Key Takeaways
- The "Onsite" is as much about stamina as it is about technical skill.
- Don't let a bad Round 1 affect your Round 2.
