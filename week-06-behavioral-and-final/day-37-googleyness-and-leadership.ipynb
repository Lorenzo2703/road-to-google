## 📚 Theory

Google evaluates candidates on two cultural pillars: **Googleyness** and **General Cognitive Ability (Leadership)**. 

### Googleyness
This is not about being "cool". It's about being a great teammate. It includes:
- **Thriving in Ambiguity**: Can you make progress even when you don't have all the facts?
- **Valuing Feedback**: How do you respond to criticism? Do you actively seek it out?
- **Putting the User First**: Do your decisions prioritize the end-user over personal or technical preferences?
- **Doing the Right Thing**: Integrity and ethics in engineering.

### Leadership (GCA)
You don't need a manager title to show leadership. Google looks for **Emergent Leadership**:
- Taking initiative to fix a problem nobody asked you to fix.
- Mentoring others or improving team processes.
- Taking ownership of a project and seeing it through to completion.

## 🔨 Practice: Hypothetical Leadership Questions

## 💡 Interview Pro-Tip

For hypothetical questions, show your work. Explain **why** you are making a certain decision. Mention the trade-offs between user experience, engineering debt, and business deadlines.

### Key Takeaways
- **Googleyness** is about collaboration and user empathy.
- **Emergent Leadership** is about taking initiative regardless of your role.
