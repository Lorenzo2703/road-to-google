## 📚 Theory

Hard-level problems rarely use a single pattern. They usually require combining two or more concepts you've learned. For example:
- **Dijkstra + DP**: Finding the shortest path with specific constraints.
- **Trie + DFS**: Searching for multiple words in a grid.
- **Sliding Window + Monotonic Queue**: Optimizing range-based queries.

### Strategy for Hard Problems
1. **Brute Force First**: If you can't see the optimal solution, write the brute force. It shows you understand the problem.
2. **Look for Subproblems**: Can this be broken down into a smaller version of itself? (DP/Recursion).
3. **Analyze Bottlenecks**: Is the bottleneck `O(N^2)` because of a nested loop? Can a Hash Map or a Heap bring it down to `O(N log N)`?

```python
# Solution
def trap_solution(height: list[int]) -> int:
    if not height: return 0
    
    l, r = 0, len(height) - 1
    left_max, right_max = height[l], height[r]
    res = 0
    
    while l < r:
        if left_max < right_max:
            l += 1
            left_max = max(left_max, height[l])
            res += left_max - height[l]
        else:
            r -= 1
            right_max = max(right_max, height[r])
            res += right_max - height[r]
            
    return res
```

**Problem**: Given an `m x n` board of characters and a list of strings `words`, return all words on the board. Each word must be constructed from letters of sequentially adjacent cells.

**Approach**: Build a **Trie** of all words. Then perform **DFS** on the board. Use the Trie to prune the search (if the current prefix isn't in the Trie, stop searching).
